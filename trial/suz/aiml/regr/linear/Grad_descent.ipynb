```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# sample data
X = np.array([1, 2, 3, 4, 5])
y = np.array([2.2, 2.8, 4.5, 3.7, 5.0])

# cost function surface
w0 = np.linspace(0, 3, 50)
w1 = np.linspace(0, 1.5, 50)
W0, W1 = np.meshgrid(w0, w1)
J = np.zeros_like(W0)

for i in range(W0.shape[0]):
    for j in range(W0.shape[1]):
        y_pred = W0[i,j] + W1[i,j]*X
        J[i,j] = np.mean((y - y_pred)**2)/2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W0, W1, J, cmap='viridis', alpha=0.8)
ax.set_xlabel('w0'); ax.set_ylabel('w1'); ax.set_zlabel('Cost')
plt.show()



```
