```python
from ipywidgets import interact, FloatSlider, IntSlider
import numpy as np
import matplotlib.pyplot as plt

# sample data
# X = np.array([1, 2, 3, 4, 5])
# y = np.array([2.2, 2.8, 4.5, 3.7, 5.0])
X = np.array([2, 3, 5, 6])
y = np.array([50, 60, 70, 70])
w0_min = 100
w1_min = 100
mse_min = 100

# Best reported:
# w0=39, w1=6, mse=9
# w0=43, w1=5, mse=6.5

def plot_line(w0, w1):
    plt.scatter(X, y)
    plt.xlim(-1, 10)
    plt.ylim(0, 90)
    y_pred = w0 + w1 * X
    plt.grid(True, which='both', linestyle='--', linewidth=0.7, color='gray', alpha=0.7)
    plt.plot(X, y_pred, color='red')
    mse = np.mean((y - y_pred)**2)
    # if mse < mse_min * 1.0:
    #     mse_min = mse
    #     w1_min = w1
    #     w0_min = w0

    plt.title(f'y = {w0:.2f} + {w1:.2f}x | MSE = {mse:.3f}')
    # plt.title(f'y = {w0:.2f} + {w1:.2f}x | MSE = {mse:.3f} '
    #           f'| MIN w0={w0_min:.3f} w1={w1_min:.3f} MSE_min={mse_min:.3f}')
    plt.show()

interact(plot_line,
         w0=IntSlider(min=0, max=55, step=1, value=5),
         w1=FloatSlider(min=-5, max=20, step=1, value=1))
```
