```python
import torch
import numpy as np
from numpy.random import random
from torch.nn import ReLU as relu
from torch.nn import ReLU as gelu

row = 3
col = 5
l1 = [i * (-1)**i for i in range(row * col)]
print(l1)
matrix = np.reshape(l1, (row, col))
t1 = torch.tensor(matrix, dtype=torch.float32)

# Error
#print(relu(t1))
print("T1: ", t1)
print("RELU(T1): ", relu()(t1))



#relu()
```

```python
import torch.nn as nn
gelu = nn.GELU()
print("GELU(T1): ", gelu(t1))
```

```python
sf = nn.Softmax(dim=-100)
print("SF_MAX(T1): ", sf(t1))
```
