```python
from pandas import DataFrame as DF
import numpy as np
import pandas as pd
from sympy.codegen.ast import float32

```

#  Load from in-memory data

```python
data = {
    'col_1' : [1, 2, 3],
    'col_2' : [10, 20, 30]
}

from_mem = DF(data, dtype=float)
print("Head: ", from_mem.head())
print("Tail: ", from_mem.tail())
print("shape: ", from_mem.shape)
print("\nKeys: ", from_mem.keys().values)
print("Values: ", from_mem.values)
print("\nDTypes: ", from_mem.dtypes)
print("\nInfo: ", from_mem.info())
print("\nDescribe: ", from_mem.describe())

```

```python
# Select a Column
col_1 = from_mem['col_1']
col_2 = from_mem['col_2']
print("\nColumn 'Col_1': \n", col_1.head())
print("\nColumn 'Col_2': \n", col_2.head())

# Select a subset of columns (use Array with array) - swapped the column order
selected = from_mem[['col_2', 'col_1']]
print("\nselected: \n", selected.head())
```

# Read from CSV as a DF

```python
df_sales = pd.read_csv('Product_sales.csv')
print("\ncsv type: \n", type(df_sales), "\nRowCount: ", df_sales['Id'].count())
print("\ncsv Head: \n", df_sales.head())
print("\ncsv Tail: \n", df_sales.tail())

print(df_sales.count())
```

### Shape and Columns

```python
#less typing
df = df_sales
print("dataframe shape: ", df.shape)

col_prod_cat = df['Product Category']
print("Prod Cat Column-shape: ", col_prod_cat.shape)
print("Unique value counts of col ProdCat: ", col_prod_cat.value_counts())
print("Prod Cat (size): ", col_prod_cat.size)
print("Prod Cat (size): ", col_prod_cat.max)
```
